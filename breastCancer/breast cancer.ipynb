```python
# M = malignant (cancerous), B = benign (non-cancerous)


# importing required liberaries
import numpy as np
import pandas as pd
import sklearn.datasets 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# loadinng data into dataframe
dataframe = pd.read_csv("breast-cancer.csv")
# print(dataframe)

dataset = sklearn.datasets.load_breast_cancer()     # pre-list of breast cancer in sklearn liberary
# print(dataset)

# dataframe = pd.DataFrame(dataset.data, columns = dataset.feature_names)      #making dataframe of pre-list of sklearn liberary
dataframe.head()   # gives output as first 5 rows

# adding taget column to the dataframe
dataframe['labels'] = dataset.target     
# added labels column to represent 0 as M (Malignant) and 1 as B (Benign) classified as integer instead of character value

dataframe.tail()       # displays last 5 rows of dataframe
```

```python
# Exclude non-numeric columns before grouping
numeric_df = dataframe.select_dtypes(include='number')

numeric_df.groupby('labels').mean()
# calculated mean off all the columns grouped by O (B) AND 1 (M) 
# 0 ke corresponding values ka alg mean and 1 ke corresponding values ka alg mean for each column 
```

```python
numeric_df
```

```python
x = numeric_df.drop(columns = 'labels', axis = 1)
y = numeric_df['labels']
```

```python
x
```

```python
y
```

```python
# Spliiting the data into training data and testing data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
# 0.2 is 20% of the data
# random_state --> randomization used, resulting in the same splits of the data.
```

```python
x.shape
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
y_test
```

```python
# model = LogisticRegression()
model = LogisticRegression(max_iter = 1000)
# added param 'max_iter' because by default maximum iteration allowed is 100 which is not sufficient for our data.
```

```python
# training logistic regression model using training data
model.fit(x_train, y_train)
# fit function is used to fit data to our model (logistic regression model)
```

```python
# accuracy on training data
x_train_prediction = model.predict(x_train)
# predict function is used to predict the lables (y) according to the value of x 
training_data_accuracy = accuracy_score(y_train, x_train_prediction)
# y_train --> true value and x_train_prediction --> predicted value by the model
```

```python
# printing the accuracy on the training data
training_data_accuracy
```

```python
# accuracy scorre in terms of test model (x_test and y_test)
x_test_prediction = model.predict(x_test)
# predict function is used to predict the lables (y) according to the value of x 
testing_data_accuracy = accuracy_score(y_test, x_test_prediction)
# y_test --> true value and x_test_prediction --> predicted value by the model
```

```python
# printing the accuracy on the testing data
testing_data_accuracy
```
